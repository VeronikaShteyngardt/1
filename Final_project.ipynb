```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
import pprint
import scipy 
import scipy.sparse as sp
import networkx as nx
import sklearn as sk
from sklearn.cluster import SpectralClustering as SC
from sklearn.preprocessing import StandardScaler


pkl_file1 = open('A.pkl', 'rb')
pkl_file2 = open('labeled_nodes.pkl', 'rb')
pkl_file3 = open('i2t.pkl', 'rb')
pkl_file4 = open('t2i.pkl', 'rb')
A = pkl.load(pkl_file1)
d = pkl.load(pkl_file2)
data3 = pkl.load(pkl_file3)
data4 = pkl.load(pkl_file4)
pkl_file1.close()
pkl_file2.close()
pkl_file3.close()
pkl_file4.close()
```

```python

n = A.shape[0]
d0 = np.zeros(n)
d0[:] = d.todense().flatten()
d = d0.copy()
d += d0.clip(0) * (np.sum(d0.clip(0) - d0)/np.sum(d0.clip(0)) - 1)

```

```python

ones_n = np.ones(n)
id_n = sp.eye(n, dtype=np.int8, format='csr')
A0 = id_n + A
F = A0.dot(sp.diags(1/A0.T.dot(ones_n), 0, format='csc'))
B = A0.T.dot(sp.diags(1/A0.dot(ones_n), 0, format='csc'))
print F.T.dot(ones_n)
print B.T.dot(ones_n)
```

```python

def reprank(F, B, d):
    alpha1, alpha2, alpha3 = 0.1, 0.5, 0.9
    t = d
    for i in xrange(50):
        t = alpha1 * F.dot(t.clip(0)) + alpha2 * B.dot(t-t.clip(0)) + alpha3 * d
    return t
```

```python

t = reprank(F, B, d)
```

```python

flag = t > 3
Aplus = A[:, flag][flag, :]
print flag.shape
Aplus = Aplus.todense()
```
